```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = pd.read_csv('/Users/zoephilipiyawa/PycharmProjects/pythonProject2/Titanic-Dataset.csv')

titanic.head()
```

```python
# how does survival vary by sex
plt.figure(figsize = (8,6))
sns.countplot(x = 'Sex', hue = 'Survived', data = titanic)
plt.title('Survival Rate by Gender')
plt.xlabel('Sex')  
plt.ylabel('Number of Survival')  

```

```python

# how survival vary by class
sns.barplot(x = 'Pclass', y ='Survived', data = titanic, hue = 'Sex')
plt.title('Survival Rate by Class')
plt.xlabel('Class')  
plt.ylabel('Number of Survival')  
```

```python
# how does survival rate vary by number of sibling ?
sns.countplot(x = 'SibSp', data = titanic, hue = 'Survived')
plt.title('Survival Rate based on Number of Sibling')
plt.xlabel('Number of Sibling')  
plt.ylabel('Number of Survival')  
```

```python
# how does average fare affect survival
fare = titanic.groupby('Survived')['Fare'].mean()
print(fare)
fare.plot(kind = 'bar', color = ['orange', 'blue'])
plt.xlabel('Survival')
plt.ylabel('Fare')
plt.title('Survival Rate by Avg Fare')
```

```python
age = titanic.groupby('Survived')['Age'].mean()
print(age)
age.plot(kind = 'bar', color = ['orange', 'blue'])
plt.xlabel('Survival')
plt.ylabel('Age')
plt.title('Survival Rate By Avg Age')
```

```python
# survival rate for people based on number of parents.
plt.figure(figsize = (8,6))
sns.countplot(x = 'Parch', hue = 'Survived', data = titanic)
plt.title('Survival Rate by Number of Parent')
plt.xlabel('Number of Parents')  
plt.ylabel('Number of Survival')  
```
